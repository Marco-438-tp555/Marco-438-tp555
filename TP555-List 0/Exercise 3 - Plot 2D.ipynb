```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

#Number of Elements
M = 1000

x = np.random.rand(M) # Uniform Distribution
w = np.random.randn(M) # Gaussian Distribution
#y = [1000,1]

y = 1.2 + 2.3*x + 10*w

fig1, axis = plt.subplots()
#axis.plot(x, y)

#Plot points
plt.scatter(x, y, color = "blue")

axis.set(xlabel='x', ylabel='y' , title='Equation: y = 1.2 + 2.3*x + 10*w')
plt.grid()

fig1.savefig("Exercise 3")

plt.show()
```
