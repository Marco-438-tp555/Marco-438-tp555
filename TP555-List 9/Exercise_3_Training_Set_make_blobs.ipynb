```python
# Import all necessary libraries.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
# Create training dataset
X, y = make_blobs(n_samples=150, n_features=2, centers=5, cluster_std=1.0, shuffle=True, random_state=42)
```

## Show Training data set

```python
# Data Ploting
fig = plt.figure(figsize=(10,10))
plt.scatter(X[:, 0], X[:,1], s = 100, color = 'blue', label = 'Data Training')
plt.xlabel('$x_1$', fontsize=14)
plt.ylabel('$x_2$', fontsize=14)
plt.legend(loc="lower right")
plt.grid()
#plt.savefig('Data_plot.png', dpi=600)
plt.show()
```

## KMeans instance

```python
# Create and instance of KMeans Cluster 
km = KMeans(n_clusters=4)
```

## Training Model and print great values

```python
# Training
km.fit(X)

# Great Centroids
for i in range(0, 4):
    print('Centroid %d: (%1.2f, %1.2f)' % (i, km.cluster_centers_[i][0], km.cluster_centers_[i][1]))
```

## Iterations to training of model

```python
# Number of iterations required to train the model
print('Number of Iterations:', km.n_iter_)
```

## Show Training data - Cluster and Centroid

```python
# Figura com os resultados
fig = plt.figure(figsize=(10,10))
plt.scatter(X[:, 0], X[:,1], s = 100, c = km.labels_)
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], s = 300, c = 'green', label = 'Centroids')
plt.title('Training data with Clusters and Centroids')
plt.xlabel('$x_1$', fontsize=14)
plt.ylabel('$x_2$', fontsize=14)
plt.legend(loc="lower right")
plt.grid()
#plt.savefig('Data_separated_by_clusters.png', dpi=600)
plt.show()
```
